```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

```

# **DATA PREPARATION**

```python
#PS1 TO PS6

temp=[]
for i in ["PS1","PS2","PS3","PS4","PS5","PS6"]:
    dataPS=np.genfromtxt( './Hydraulics/''/'+i+'.txt')
    temp.append(dataPS)
ps1,ps2,ps3,ps4,ps5,ps6=temp
ps1_arr=ps1.mean(axis=1)
ps2_arr=ps2.mean(axis=1)
ps3_arr=ps3.mean(axis=1)
ps4_arr=ps4.mean(axis=1)
ps5_arr=ps5.mean(axis=1)
ps6_arr=ps6.mean(axis=1)
```

```python
data_PS={'PS1':ps1_arr,'PS2':ps2_arr,'PS3':ps3_arr,'PS4':ps4_arr,'PS5':ps5_arr,'PS6':ps6_arr}
df=pd.DataFrame(data_PS)
```

```python
#EPS1

data_e1=np.genfromtxt( './Hydraulics/EPS1.txt')
eps1_arr=data_e1.mean(axis=1)
df['EPS1']=eps1_arr
```

```python
#FS1 AND FS2

temp=[]
for i in ['FS1','FS2']:
    dataFS=np.genfromtxt('./Hydraulics/''/'+i+'.txt')
    temp.append(dataFS)
fs1,fs2=temp
fs1_arr=fs1.mean(axis=1)
fs2_arr=fs2.mean(axis=1)
df['FS1']=fs1_arr
df['FS2']=fs2_arr
```

```python
#TS1 TO TS4

temp=[]
for i in ['TS1','TS2','TS3','TS4']:
    dataTS=np.genfromtxt('./Hydraulics/''/'+i+'.txt')
    temp.append(dataTS)
ts1,ts2,ts3,ts4=temp
ts1_arr=ts1.mean(axis=1)
ts2_arr=ts2.mean(axis=1)
ts3_arr=ts3.mean(axis=1)
ts4_arr=ts4.mean(axis=1)
df['TS1']=ts1_arr
df['TS2']=ts2_arr
df['TS3']=ts3_arr
df['TS4']=ts4_arr
```

```python
#VS1

data_v1=np.genfromtxt('./Hydraulics/VS1.txt')
vs1_arr=data_v1.mean(axis=1)
df['VS1']=vs1_arr
```

```python
#REMAINING SENSOR DATA

temp=[]
for i in ['CE','CP','SE']:
    dataRemain=np.genfromtxt('./Hydraulics/''/'+i+'.txt')
    temp.append(dataRemain)
ce,cp,se=temp
ce_arr=ce.mean(axis=1)
cp_arr=cp.mean(axis=1)
se_arr=se.mean(axis=1)
df['CE']=ce_arr
df['CP']=cp_arr
df['SE']=se_arr
```

```python
#TARGET(PROFILE.TXT)

profile_11=np.genfromtxt('./Hydraulics/profile.txt')
df_temp=pd.DataFrame(profile_11,columns=['Cooler condition','Valve condition','Internal pump leakage','Hydraulic accumulator / bar','Stable flag'])
df_temp
```

```python
#FINAL DATASET

df_final=pd.concat([df,df_temp],axis=1)
df_final.head()
```
